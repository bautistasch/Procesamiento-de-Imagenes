```python
import numpy as np
from scipy.ndimage import rotate

def fspecial(type, arg1=None, arg2=None):
    if not isinstance(type, str):
        raise ValueError("fspecial: first argument must be a string")

    type = type.lower()

    if type == "average":
        if arg1 is not None and np.all(np.isreal(arg1)) and np.prod(arg1.shape) <= 2:
            fsize = arg1
        else:
            fsize = 3
        f = np.ones((fsize, fsize))
        f /= np.sum(f)

    elif type == "disk":
        if arg1 is not None and np.all(np.isreal(arg1)) and np.isscalar(arg1):
            radius = arg1
        else:
            radius = 5
        x, y = np.meshgrid(np.arange(-radius, radius+1), np.arange(-radius, radius+1))
        r = np.sqrt(x**2 + y**2)
        f = (r <= radius).astype(float)
        f /= np.sum(f)

    elif type == "gaussian":
        if arg1 is not None and np.all(np.isreal(arg1)):
            if np.isscalar(arg1):
                hsize = (arg1, arg1)
            elif isinstance(arg1, (list, tuple)) and len(arg1) == 2:
                hsize = arg1
            else:
                raise ValueError("fspecial: second argument must be a scalar or a vector of two scalars")
        else:
            hsize = (3, 3)
        
        if arg2 is not None and np.all(np.isreal(arg2)) and np.isscalar(arg2):
            sigma = arg2
        else:
            sigma = 0.5
        
        h1, h2 = hsize[0] - 1, hsize[1] - 1
        x, y = np.meshgrid(np.arange(0, h2+1), np.arange(0, h1+1))
        x = x - h2/2
        y = y - h1/2
        gauss = np.exp(-(x**2 + y**2) / (2 * sigma**2))
        f = gauss / np.sum(gauss)

    elif type == "laplacian":
        if arg1 is not None and np.isscalar(arg1):
            alpha = arg1
            if alpha < 0 or alpha > 1:
                raise ValueError("fspecial: second argument must be between 0 and 1")
        else:
            alpha = 0.2
        f = (4 / (alpha + 1)) * np.array([[alpha/4, (1-alpha)/4, alpha/4],
                                         [(1-alpha)/4, -1, (1-alpha)/4],
                                         [alpha/4, (1-alpha)/4, alpha/4]])

    elif type == "log":
        if arg1 is not None and np.all(np.isreal(arg1)):
            if np.isscalar(arg1):
                hsize = (arg1, arg1)
            elif isinstance(arg1, (list, tuple)) and len(arg1) == 2:
                hsize = arg1
            else:
                raise ValueError("fspecial: second argument must be a scalar or a vector of two scalars")
        else:
            hsize = (5, 5)
        
        if arg2 is not None and np.all(np.isreal(arg2)) and np.isscalar(arg2):
            sigma = arg2
        else:
            sigma = 0.5
        
        h1, h2 = hsize[0] - 1, hsize[1] - 1
        x, y = np.meshgrid(np.arange(0, h2+1), np.arange(0, h1+1))
        x = x - h2/2
        y = y - h1/2
        gauss = np.exp(-(x**2 + y**2) / (2 * sigma**2))
        f = ((x**2 + y**2 - 2 * sigma**2) * gauss) / (2 * np.pi * sigma**6 * np.sum(gauss))

    elif type == "motion":
        if arg1 is not None and np.all(np.isreal(arg1)):
            len_ = int(arg1)
        else:
            len_ = 9
        
        if len_ % 2 == 1:
            sze = (len_, len_)
        else:
            sze = (len_+1, len_+1)
        
        if arg2 is not None and np.all(np.isreal(arg2)):
            angle = arg2
        else:
            angle = 0
        
        f = np.zeros(sze)
        f[len_//2, 0:len_] = 1
        f = rotate(f, angle, reshape=False)
        f /= np.sum(f)

    elif type == "prewitt":
        f = np.array([[1, 1, 1],
                      [0, 0, 0],
                      [-1, -1, -1]])

    elif type == "sobel":
        f = np.array([[1, 2, 1],
                      [0, 0, 0],
                      [-1, -2, -1]])

    elif type == "kirsch":
        f = np.array([[3, 3, 3],
                      [3, 0, 3],
                      [-5, -5, -5]])

    elif type == "unsharp":
        if arg1 is not None and np.isscalar(arg1):
            alpha = arg1
            if alpha < 0 or alpha > 1:
                raise ValueError("fspecial: second argument must be between 0 and 1")
        else:
            alpha = 0.2
        
        f = (1 / (alpha + 1)) * np.array([[-alpha, alpha-1, -alpha],
                                         [alpha-1, alpha+5, alpha-1],
                                         [-alpha, alpha-1, -alpha]])

    else:
        raise ValueError("fspecial: filter type '{}' is not supported".format(type))

    return f
```

```python
# Filters

filters = {
    'unsharp': fspecial('unsharp'),
    'disk': fspecial('disk'),
    'array': np.array([
        [0  , 1/6, 0  ],
        [1/6, 1/3, 1/6],
        [0  , 1/6, 0  ],
    ]),
}
```

```python
def show(title, h):
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
    fig.suptitle(title, y=0.8)
    n = 256
    big = np.zeros((n, n))
    big[n // 2, n // 2] = 1
    h1 = scipy.signal.convolve2d(big, h, mode="same")
    ax1.set_title("Conv")
    ax1.imshow(h1)

    S = scipy.fft.fft2(h1)
    SM = np.abs(S)
    img = scipy.fft.fftshift(SM / np.max(SM))
    ax2.set_title("Spect")
    ax2.imshow(img)

    IMd = np.log(1 + np.abs(SM))
    img = scipy.fft.fftshift(IMd / np.max(SM))
    ax3.set_title("Spect Log")
    ax3.imshow(img)

    ax1.axis('off')
    ax2.axis('off')
    ax3.axis('off')
    plt.tight_layout()
    plt.show()

show('HI pass', filters['unsharp'])
show('LOW pass', filters['disk'])
show('array', filters['array'])
```

### Funcion upsampling

```python
def upsampling(I,m):
# Upsample the square image I by a factor of m
  N, M=np.shape(I)
  s=(m*N,m*N)
  Iup = np.zeros(s)
  # Expand input image
  for i in range(N):
    for j in range(N):
        Iup[m*(i-1)+1,m*(j-1)+1]=I[i,j]
# Ideal filter
  N, M=np.shape(Iup)
  w=1/m
  F=scipy.fft.fftshift(scipy.fft.fft2(Iup))
  for i in range(N):
      for j in range(N):
          r2=(i-round(N/2))**2+(j-round(N/2))**2
          if (r2>round((N/2*w)**2)):
            F[i,j]=0
  Iup=(m*m)*abs(scipy.fft.ifft2(scipy.fft.fftshift(F)))
  return Iup
```

### Funcion DownSampling

```python
def downsampling(I,m,filter):
# Downsample the square image I by a factor of m
  N, M=np.shape(I)
                    # Apply ideal filter
  w=1/m;
  F=scipy.fft.fftshift(scipy.fft.fft2(I));
  if filter=='FILTER_ON':
    for i in range(N):
        for j in range(N):
            r2=(i-round(N/2))**2+(j-round(N/2))**2;

            if (r2>round((N/2*w)**2)):        #ojo w=1/m
              F[i,j]=0



  Idown=np.real(scipy.fft.ifft2(scipy.fft.fftshift(F)));
# Now downsample

  Idown = cv2.resize(Idown, dsize=(int(N/m), int(N/m)), interpolation=cv2.INTER_NEAREST)
  return Idown
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal, fft, linalg
from PIL import Image
import cv2
```

La estructura ser√°, cargar las 2 imagenes, hallar las respectivas transformadas, hallar la norma cuadrada de la convolucion entre ambas y luego sacar R, para despues aplicar IDFT y quedarnos con max(r)

```python
ia = cv2.imread('dog1ss.jpg',cv2.IMREAD_GRAYSCALE)
ib = cv2.imread('dog2ss.jpg',cv2.IMREAD_GRAYSCALE)
plt.figure(1)
plt.imshow(ia,cmap='gray', vmin=0, vmax=255)
plt.title('ia')
plt.axis('off')
plt.show(block=False) # Show the plot in non-blocking mode (i.e continue)
plt.figure(2)
plt.imshow(ib,cmap='gray', vmin=0, vmax=255)
plt.title('ib')
plt.axis('off')
plt.show(block=False) # Show the plot in non-blocking mode (i.e continue)
```

```python
print(ia.shape)  #Vemos el tama√±o de las imagenes
```

```python
Ia = fft.fft2(ia)   #Transformamos ambas imagenes,
Ib = fft.fft2(ib)
Ia = np.real(Ia)    #Y nos quedamos solo con la parte real (ambas se√±ales son reales y buscamos minimizar el ruido num√©rico)
Ib = np.real(Ib)
```

Realizamos el producto Hadamard (elemento a elemento) entre las matrices

```python
iaxib = Ia*Ib
```

```python
minorm=linalg.norm(iaxib)   # Norma 1 del producto ia x ib
```

```python
R = iaxib / minorm      ## Correlaci√≥n normalizada
r = fft.ifft2(R)        ## IDFT[R(u,v)] = r(x,y)
```

```python
print("Vemos el valor de r.shape: ",r.shape,"\nL√≥gicamente se mantuvo la dimension de las im√°genes en la FFT y en el producto punto")
#Buscamos en que indice se enceuntra el valor maximo, siendo que mis indices son x e y, se maximizan en DeltaX / DeltaY
deltaY,deltaX = np.unravel_index(r.argmax(), r.shape)
deltaYoffs = deltaY-r.shape[0]            #Como el desplazamiento en Y es negativo, se interpreta el wrap-around
print("Los valores de deltaX y deltaY resultan \nDeltaX:",deltaX,"\nDeltaY:",deltaYoffs)
```

En realidad el deltaY es $\Delta Y = 268px-276px = -8px$, ya que lo que se aprecia es un corrimiento en el eje Y hacia las coordinadas negativas.
Como el corrimiento en X es en sentido positivo, $\Delta X = 107px$ es correcto
